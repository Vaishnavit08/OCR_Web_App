```python
from PIL import Image
from byaldi import RAGMultiModalModel
from transformers import Qwen2VLForConditionalGeneration, AutoProcessor
from qwen_vl_utils import process_vision_info
import torch
```

```python
# Load the pre-trained ColPali model
RAG = RAGMultiModalModel.from_pretrained("vidore/colpali")
```

```python
# Example of how you can load an image
image_path = "/content/Sentence Pattern.jpg"  # Replace with your image file path
image = Image.open(image_path)
```

```python
# Index the image using Byaldi
RAG.index(
    input_path=image_path,
    index_name="image_index",  # index will be saved at index_root/index_name/
    store_collection_with_index=False,
    overwrite=True
)
```

```python
# Query and search (assuming multi-modal inputs)
text_query = "what is text"
results = RAG.search(text_query, k=1)
```

```python
# Load the Qwen2-VL model and processor
model = Qwen2VLForConditionalGeneration.from_pretrained("Qwen/Qwen2-VL-2B-Instruct",
                                                        trust_remote_code=True, torch_dtype=torch.bfloat16).cuda().eval()

processor = AutoProcessor.from_pretrained("Qwen/Qwen2-VL-2B-Instruct", trust_remote_code=True)
```

```python
# Using the results obtained from the search
image_index = results[0]["page_num"] - 1  # Using the page number returned by Byaldi
messages = [
    {
        "role": "user",
        "content": [
            {
                "type": "image",
                "image": image,  # Image object from PIL
            },
            {"type": "text", "text": text_query},
        ],
    }
]
```

```python
# Prepare the input for the model
text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
```

```python
# Process image information
image_inputs, video_inputs = process_vision_info(messages)
inputs = processor(
    text=[text],
    images=image_inputs,
    videos=video_inputs,
    padding=True,
    return_tensors="pt"
)
inputs = inputs.to("cuda")
```

```python
# Generate the output (extracted text)
generated_ids = model.generate(**inputs, max_new_tokens=50)
generated_ids_trimmed = [
    out_ids[len(in_ids):] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
]
output_text = processor.batch_decode(
    generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
)
```

```python
import gradio as gr
from PIL import Image, ImageOps
import pytesseract
import re
import json
import time

# Function to highlight multiple keywords in different colors
def highlight_multiple_keywords(text, keywords):
    colors = ['#FF5733', '#33FF57', '#3357FF', '#FF33A1', '#33FFF6']  # Predefined colors for keywords
    for i, keyword in enumerate(keywords):
        color = colors[i % len(colors)]  # Rotate through the color list
        text = re.sub(rf"\b({keyword})\b", rf"<span style='color: {color}; font-weight: bold;'>\1</span>", text, flags=re.IGNORECASE)
    return text

# OCR process function with image preprocessing and language selection
def ocr_process(image, query_text, language, preprocess):
    start_time = time.time()  # Start time for efficiency tracking

    # Step 1: Open the uploaded image
    image = Image.open(image)

    # Step 2: Apply image preprocessing if selected
    if preprocess == "Grayscale":
        image = ImageOps.grayscale(image)
    elif preprocess == "Resize (50%)":
        image = image.resize((image.width // 2, image.height // 2))

    # Step 3: Configure Tesseract based on language selection
    tesseract_config = "--psm 6"
    if language == "English":
        tesseract_config += " -l eng"
    elif language == "Hindi":
        tesseract_config += " -l hin"
    else:
        tesseract_config += " -l eng+hin"

    # Step 4: Extract text using Tesseract OCR
    extracted_text = pytesseract.image_to_string(image, config=tesseract_config)

    # Step 5: Search for multiple keywords (split by commas)
    keywords = [kw.strip() for kw in query_text.split(',')]  # Split input by commas for multiple keywords
    highlighted_text = highlight_multiple_keywords(extracted_text, keywords)

    # Find matches and positions
    matches_info = []
    for keyword in keywords:
        matches = re.finditer(rf"\b{keyword}\b", extracted_text, re.IGNORECASE)
        match_positions = [match.start() for match in matches]
        matches_info.append(f"Keyword '{keyword}' found at positions: {match_positions}" if match_positions else f"Keyword '{keyword}' not found.")

    search_result = "\n".join(matches_info)

    # Step 6: Create a downloadable JSON file with extracted text and search results
    result_data = {
        'extracted_text': extracted_text,
        'search_query': query_text,
        'search_result': search_result
    }
    result_json = json.dumps(result_data, indent=4)

    # Save the result to a file
    with open("ocr_result.json", "w") as file:
        file.write(result_json)

    # Step 7: Save extracted text as plain text
    with open("ocr_extracted_text.txt", "w") as file:
        file.write(extracted_text)

    # Step 8: Efficiency improvement - time taken for OCR process
    extraction_time = time.time() - start_time

    return highlighted_text, search_result, "ocr_result.json", "ocr_extracted_text.txt", f"Time taken for extraction: {extraction_time:.2f} seconds"

# Define Gradio interface
interface = gr.Interface(
    fn=ocr_process,
    inputs=[
        gr.Image(type="filepath", label="Upload an image"),  # Accepts image file
        gr.Textbox(label="Enter search keywords (comma-separated for multiple)"),  # Multiple keyword input
        gr.Dropdown(choices=["English", "Hindi", "English + Hindi"], label="Choose OCR Language"),  # Language selection
        gr.Radio(choices=["None", "Grayscale", "Resize (50%)"], label="Image Preprocessing")  # Image preprocessing options
    ],
    outputs=[
        gr.HTML(label="OCR Results with highlighted search text"),  # Display highlighted text with HTML
        gr.Textbox(label="Search Result"),  # Show if keyword is found and its position
        gr.File(label="Download Extracted Text as JSON"),  # Download the output as JSON
        gr.File(label="Download Extracted Text as .txt"),  # Download extracted text as plain text
        gr.Textbox(label="Efficiency (Time taken)")  # Show extraction time for efficiency
    ],
    title="Multilingual OCR & Advanced Keyword Search Application",
    description="Upload an image, preprocess it, extract text in English or Hindi, search for multiple keywords, and download results in JSON or plain text.",
    theme="default"
)

# Launch the interface in Colab
interface.launch()
```
